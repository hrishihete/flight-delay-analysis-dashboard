# Week 2 Data Cleaning

Project: Flight Delay Analysis & Dashboard

Goal: clean BTS flight delay data, handle nulls in delay fields, and filter columns for EDA/Tableau.

## Cleaning Rules

- Keep relevant airline, airport, time, delay, cancellation, diversion, distance, and delay-cause fields.
- Fill delay-cause nulls with `0` because BTS leaves these blank unless a qualifying arrival delay exists.
- Fill missing cancellation codes with `Not Cancelled`.
- Add flags for original departure and arrival delay nulls.
- Track completed flights separately from cancelled or diverted flights.
